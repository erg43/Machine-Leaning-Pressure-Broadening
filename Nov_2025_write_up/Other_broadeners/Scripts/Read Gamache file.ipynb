```python
# Code for paper 29

infile = '/Users/elizabeth/Downloads/1-s2.0-S0022407319301396-mmc1 2/H2O-CO2_2019.txt'
H2OCO2 = open(infile)
nt = 0


nt_data = []
mol = []
iso = []
freq = []
S = []
A = []
E_lower = []
g_CO2 = []
c1 = []
n1 = []
c2 = []
n2 = []
shift = []
d1 = []
q1 = []
d2 = []
q2 = []
gself = []
V_upper = []
V_lower = []
Qrot_up = []
Qrot_lo = []
Ie_F = []
Ie_S = []
Ie_gCO2 = []
Ie_gself = []
Ie_nCO2 = []
Ie_dCO2 = []
IRef_F = []
IRef_S = []
IRef_g_CO2 = []
IRef_gself = []
IRef_n = []
IRef_shift_CO2 = []
gj_up = []
gj_lo = []

for line in H2OCO2:
    nt = nt + 1
    nt_data.append(nt)
    line = line.rstrip()
    mol.append(line[0:2])
    iso.append(line[2:3])
    freq.append(float(line[3:15]))
    S.append(float(line[15:25]))
    A.append(float(line[25:35]))
    E_lower.append(float(line[35:45]))
    g_CO2.append(float(line[45:52]))
    c1.append(float(line[52:62]))
    n1.append(float(line[62:72]))
    c2.append(float(line[72:82]))
    n2.append(float(line[82:92]))
    shift.append(float(line[92:100]))
    d1.append(float(line[100:110]))
    q1.append(float(line[110:120]))
    d2.append(float(line[120:130]))
    q2.append(float(line[130:140]))
    gself.append(float(line[140:145]))
    V_upper.append(line[145:160])
    V_lower.append(line[160:175])
    Qrot_up.append(line[175:190])
    Qrot_lo.append(line[190:205])
    Ie_F.append(line[205:206])
    Ie_S.append(line[206:207])
    Ie_gCO2.append(line[207:208])
    Ie_gself.append(line[208:209])
    Ie_nCO2.append(line[209:210])
    Ie_dCO2.append(line[210:211])
    IRef_F.append(line[211:213])
    IRef_S.append(line[213:215])
    IRef_g_CO2.append(line[215:217])
    IRef_gself.append(line[217:219])
    IRef_n.append(line[219:221])
    IRef_shift_CO2.append(line[221:223])
    gj_up.append(float(line[224:231]))
    gj_lo.append(float(line[231:238]))
    
H2OCO2.close()

dataframe = pd.DataFrame({'nt': nt_data,
'mol': mol, 
'iso': iso, 
'freq': freq, 
'S': S, 
'A': A, 
'E_lower': E_lower, 
'g_CO2': g_CO2, 
'c1': c1, 
'n1': n1, 
'c2': c2, 
'n2': n2, 
'shift': shift, 
'd1': d1, 
'q1': q1, 
'd2': d2, 
'q2': q2, 
'gself': gself, 
'V_upper': V_upper, 
'V_lower': V_lower, 
'Qrot_up': Qrot_up, 
'Qrot_lo': Qrot_lo, 
'Ie_F': Ie_F, 
'Ie_S': Ie_S, 
'Ie_gCO2': Ie_gCO2, 
'Ie_gself': Ie_gself, 
'Ie_nCO2': Ie_nCO2, 
'Ie_dCO2': Ie_dCO2, 
'IRef_F': IRef_F, 
'IRef_S': IRef_S, 
'IRef_g_CO2': IRef_g_CO2, 
'IRef_gself': IRef_gself, 
'IRef_n': IRef_n, 
'IRef_shift_CO2': IRef_shift_CO2, 
'gj_up': gj_up, 
'gj_lo': gj_lo})

dataframe = dataframe[['iso', 'freq', 'g_CO2', 'c1', 'n1', 'c2', 'n2', 'V_upper', 'V_lower', 'Qrot_up', 'Qrot_lo', 'Ie_gCO2']]

dataframe = dataframe.rename(columns={'iso':'local_iso_ID', 'freq':'nu', 'g_CO2':'gamma_CO2', 'Ie_gCO2':'gamma_CO2-err'})

dataframe = dataframe.rename(columns={'paper_in_my_list':'paper_number_in_my_list'})

dataframe[['Jpp','Kapp', 'Kcpp']] = dataframe['Qrot_lo'].str.split('\s+',expand=True)[[1, 2, 3]]
dataframe[['J','Ka', 'Kc']] = dataframe['Qrot_up'].str.split('\s+',expand=True)[[1, 2, 3]]
dataframe = dataframe.drop(['Qrot_up', 'Qrot_lo'], axis=1)

dataframe['Temperature'] = 296
dataframe['profile'] = 'Voigt'
dataframe['gamma_uncertainty'] = 0.05*dataframe['gamma_CO2']

dataframe.to_csv('H2O_by_CO2_paper_29.csv')
```
