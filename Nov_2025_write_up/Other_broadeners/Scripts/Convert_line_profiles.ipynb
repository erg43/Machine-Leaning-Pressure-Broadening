```python
def hartmann_tran(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0):
    F_ht = (1/np.pi) * ( A(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0) / (1 - ((nu_VC - (eta * (C_0(Gamma_0, Delta_0) - 3*C_2(Gamma_2, Delta_2)/2))) * A(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)) + ((eta*C_2(Gamma_2, Delta_2)/(nu_a0(nu_0, Gamma_D)**2)) * B(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)))).real
    return F_ht
```

```python
def A(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0):
    A = ((np.sqrt(np.pi) * c / (nu_0 * nu_a0(nu_0, Gamma_D)))
                * (w(complex(0, 1) * Z_minus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0))
                - w(complex(0, 1) * Z_plus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0))))
    return A
```

```python
def B(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0):
    B = (nu_a0(nu_0, Gamma_D)**2 / C_2_tilde(eta, Gamma_2, Delta_2)) * (-1     
                + ((np.sqrt(np.pi)/(2*np.sqrt(Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2)))) * (1 - Z_minus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)**2) * w(complex(0, 1) * Z_minus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)))
                - ((np.sqrt(np.pi)/(2*np.sqrt(Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2)))) * (1 - Z_plus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)**2) * w(complex(0, 1) * Z_plus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0))))
    return B
```

```python
def w(z):
    w_list = []
    for value in z:
        w = complex(0, 1) * (1/np.pi) * quad(integrand, -np.inf, np.inf, args=(value))[0]
        w_list.append(w)
    w_arr = np.array(w_list)
    return w_arr
```

```python
def integrand(t, z):
    int = np.exp(-(t**2)) / (z - t)
    return int
```

```python
def Z_plus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0):
    Z = (np.sqrt(
               X(nu, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)
               + Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2))
               + np.sqrt(Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2)))
    return Z
```

```python
def Z_minus(nu, Gamma_D, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0):
    Z = (np.sqrt(
               X(nu, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0)
               + Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2))
               - np.sqrt(Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2)))
    return Z
```

```python
def X(nu, Gamma_0, Delta_0, Gamma_2, Delta_2, nu_VC, eta, nu_0):
    X = (((complex(0, -1) * (nu_0 - nu))
                + C_0_tilde(eta, nu_VC, Gamma_0, Delta_0, Gamma_2, Delta_2)) 
                 / C_2_tilde(eta, Gamma_2, Delta_2))
    return X
```

```python
def Y(nu_0, Gamma_D, eta, Gamma_2, Delta_2):
    Y = (nu_0 * nu_a0(nu_0, Gamma_D)
                 / (2*c*C_2_tilde(eta, Gamma_2, Delta_2)))**2
    return Y
```

```python
def C_0_tilde(eta, nu_VC, Gamma_0, Delta_0, Gamma_2, Delta_2):
    C = ((1 - eta) * (C_0(Gamma_0, Delta_0) - 3*C_2(Gamma_2, Delta_2)/2)) + nu_VC
    return C
```

```python
def C_2_tilde(eta, Gamma_2, Delta_2):
    C = (1 - eta) * C_2(Gamma_2, Delta_2)
    return C
```

```python
def C_0(Gamma_0, Delta_0):
    C = complex(Gamma_0, Delta_0)
    return C
```

```python
def C_2(Gamma_2, Delta_2):
    C = complex(Gamma_2, Delta_2)
    return C
```

```python
def nu_a0(nu_0, Gamma_D):
    nu = (c / (np.sqrt(np.log(2)) * nu_0)) * Gamma_D
    return nu
```
