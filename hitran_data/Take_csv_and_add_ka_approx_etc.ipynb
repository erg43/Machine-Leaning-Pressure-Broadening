```python
for key, data in db.items():
    # symmetric linear
    if key in ['C2H2', 'CO2', 'H2', 'N2', 'O2', 'S2']:
        data['symmetry'] = 1
        db[key] = data

    # asymmetric linear
    if key in ['ClO', 'CO', 'CS', 'HBr', 'HCl', 'HF', 'N2O', 'NO', 'OCS', 'OH']:
        data['symmetry'] = 4
        db[key] = data

    # symmetric top
    if key in ['C2H6', 'CH3F', 'NH3', 'CH3CN', 'SO3', 'NF3']:
        data['symmetry'] = 3
        db[key] = data

    # asymmetric top
    if key in ['C2H4', 'CH3OH', 'COF2', 'H2CO', 'H2O2', 'H2O', 'HO2', 'NO2', 'O3', 'SO2', 'HONO', 'ClONO2']:
        data['symmetry'] = 2
        db[key] = data

    # spherical top
    if key in ['CH4', 'GeH4', 'SF6', 'CF4']:
        data['symmetry'] = 5
        db[key] = data
```
